# 7. frequent vs non-frequent customers

```python
#Create an order frequency flag that marks the regularity of a user’s ordering behavior according 
#to the median in the “days_since_prior_order” column
```

```python
#If the median of “days_since_prior_order” is higher than 20, 
#then the customer should be labeled a “Non-frequent customer.”
```

```python
#If the median is higher than 10 and lower than or equal to 20, 
#then the customer should be labeled a “Regular customer.”
```

```python
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
#new column added median_days_since_prior_order
ords_prods_merge2['median_days_since_prior_order'] = ords_prods_merge2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge2.head()
```

```python
#criteria added
ords_prods_merge2.loc[ords_prods_merge2['median_days_since_prior_order'] > 20, 'freq_flag'] = 'Non_Freq_customer'
ords_prods_merge2.loc[(ords_prods_merge2['median_days_since_prior_order']> 10)& (ords_prods_merge2['median_days_since_prior_order'] <=20 ), 'freq_flag'] = 'Regular_customer'
ords_prods_merge2.loc[ords_prods_merge2['median_days_since_prior_order']<=10, 'freq_flag']= 'Freq_customer'
```

```python
ords_prods_merge2['freq_flag'].value_counts()
```

```python
ords_prods_merge=ords_prods_merge2
```
